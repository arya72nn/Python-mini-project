```python
# Expense Tracker System

FILE = "expenses.txt"

def add_expense():
    date = input("Enter date (YYYY-MM-DD): ")
    category = input("Enter category (Food/Travel/Rent/etc): ")
    amount = input("Enter amount: ")

    with open(FILE, "a") as f:
        f.write(f"{date},{category},{amount}\n")

    print("Expense added successfully!")

def view_expenses():
    try:
        with open(FILE, "r") as f:
            print("\nDate       Category    Amount")
            print("--------------------------------")
            for line in f:
                date, category, amount = line.strip().split(",")
                print(date, category.ljust(10), amount)
    except FileNotFoundError:
        print("No expenses found!")

def total_expense():
    total = 0
    try:
        with open(FILE, "r") as f:
            for line in f:
                _, _, amount = line.strip().split(",")
                total += int(amount)
        print("Total Expense: â‚¹", total)
    except FileNotFoundError:
        print("No data available!")

def category_summary():
    summary = {}
    try:
        with open(FILE, "r") as f:
            for line in f:
                _, category, amount = line.strip().split(",")
                summary[category] = summary.get(category, 0) + int(amount)

        print("\nCategory-wise Expense:")
        for cat, amt in summary.items():
            print(cat, ":", amt)
    except FileNotFoundError:
        print("No data available!")

# ---------------- MAIN MENU ----------------
while True:
    print("\n===== Expense Tracker =====")
    print("1. Add Expense")
    print("2. View Expenses")
    print("3. Total Expense")
    print("4. Category Summary")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_expense()
    elif choice == "2":
        view_expenses()
    elif choice == "3":
        total_expense()
    elif choice == "4":
        category_summary()
    elif choice == "5":
        print("Thank you!")
        break
    else:
        print("Invalid choice!")
```
